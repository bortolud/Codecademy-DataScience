```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('jeopardy.csv')
df.columns = ['show_number','air_date','round','category','value','question','answer']
print(df.head())
```

```python
def word_filter(data, word_list):
    filter = lambda row: all(word.lower() in row.lower() for word in word_list)
    return data.loc[data["question"].apply(filter)]

words=['England', 'King']
word_filter(df, words)
```

```python
df['float_value']= df['value'].apply(lambda row: float(row[1:].replace(',','') if row != 'None' else 0))
#print(df.head())
filtered = word_filter(df, ["king"])
print(filtered.float_value.mean())
```

```python
def count_unique(data):
    new_data = data.groupby('answer').question.count().reset_index()
    return new_data

count_unique(filtered)
```
